# Aula 04 - Árvores binárias

**BST**: Árvore BInária de Busca <br>
**Percurso em Ordem**(Esquerda, raiz, direita): Vamos a extrema esquerda da nossa árvore, volta a raiz e vai a direita.

Exemplo:

#             7
#           /   \
#         4       9
#       /  \     /   \
#      0    5   8     13

O _percurso em ordem_ será: 0,4,5,7,8,9,13 <br>
Logo, teremos uma _ordenação crescente_ da árvore. <br><br>


**Percurso em Pré Ordem**(Raiz, esquerda, direita): Começa da raiz fazendo 'print', vai para esquerda e depois para a direita. Os 'prints' não se repetem, pois já foram contabilizados.

O _percurso em pré ordem_ será: 7,4,0,5,9,8,13 <br><br>

**Percurso em Pós Ordem**(Esquerda, direita, raiz): Vamos a extrema esquerda da nossa árvore, voi para direita e volta para raiz.


O _percurso em pós ordem_ será: 0,5,4,8,13,9,7 <br><br>

```python
# Arvore Binaria : BST
# Script em POO:

class BST:

  def __init__(self, dado=None):
    self.dado = dado
    self.esquerda = None
    self.direita = None

  def inserir(self, dado): #Metodo Inserir
    if self.dado == None: #verifica se a raiz (ou nó) esta vázio, se sim, insere o valor nela
      self.dado = dado
    else: # se não esta vazio
      if dado < self.dado: # Se o dado for menor que o valor da raiz(ou nó)
        if(self.esquerda): # menor que o head = esquerda
          self.esquerda.inserir(dado) #recursividade
        else:
          self.esquerda = BST(dado)
      else:
        if(self.direita): # maior que o head = esquerda
          self.direita.inserir(dado) #recursividade
        else:
          self.direita = BST(dado)

    # Em Comum entre os métodos abaixo:
    # 1 - Todos Recebem uma lista (lst), que começa vázia e vai sendo preenchida a medida que vai andando na árvore

  def emOrdem(self,lst): #Método que coloca os dados em ordem (esquerda, raiz, direita)
    if self.esquerda:
      self.esquerda.emOrdem(lst) # ordenando o lado esquerdo com recursividade
    lst.append(self.dado) # add na raiz
    if self.direita:
      self.direita.emOrdem(lst) #Ordenando o lado direito com recursividade
    return lst

  def preOrdem(self, lst): # (raiz,esquerda, direita)
    lst.append(self.dado) # add na raiz
    if self.esquerda:
      self.esquerda.preOrdem(lst) # Ordenando o lado esquerdo com recursividade
    if self.direita:
      self.direita.preOrdem(lst) # Ordenando o lado direito com recursividade
    return lst

  def posOrdem(self, lst): # (esquerda, direita, raiz)
    if self.esquerda:
      self.esquerda.posOrdem(lst) # Ordenando o lado esquerdo com recursividade
    if self.direita:
      self.direita.posOrdem(lst) # Ordenando o lado direito com recursividade
    lst.append(self.dado) # add na raiz
    return lst

  def emNivel(self):
    nodoAtual = self
    lst =[]
    fila=[]
    fila.insert(0,nodoAtual)
    while(len(fila) > 0):
      nodoAtual = fila.pop()
      lst.append(nodoAtual.dado)
      if nodoAtual.esquerda:
        fila.insert(0,nodoAtual.esquerda)
      if nodoAtual.direita:
        fila.insert(0,nodoAtual.direita)
    return lst


# Main:

Teste = BST() #Classe BST

Teste.inserir(7)
Teste.inserir(4)
Teste.inserir(9)
Teste.inserir(0)
Teste.inserir(5)
Teste.inserir(8)
Teste.inserir(13)

#Escopo:

#             7
#           /   \
#         4       9
#       /  \     /   \
#      0    5   8     13

print('Em ordem ', Teste.emOrdem([]))
print('Em PRÈ ordem ', Teste.preOrdem([]))
print('Em PÓS ordem ', Teste.posOrdem([]))
print('Em Nivel(Organizaçã da árvore): ', Teste.emNivel())

```

# Árvore Binária AVL

**Árvore AVL**: Chama-se AVL devido as iniciais do seus desenvolvedores (Adelson Welsk, Lands). <br>
Que é uma **árvore binária do tipo balanceada**.<br>

- o que é uma árvore desbalanceada?<br>
Quando temos uma longo ramo, apenas em um direção (esquerda) <br><br>

#             7
#           /   
#         4       
#       /       
#      0    

o que acarreta em problemas na hora de se fazer buscas. Podendo ser até reescrita com uma _lista encadeada_:<br>
- [raiz/head] = 8 -> 6 -> 4 -> 2 -> None <br>


Diante disto, uma **árvore binária do tipo balanceada** significa ter o maior número de ramos possíveis para facilitar nossa busca.

**Árvore AVL**: Tem como objetivo **melhorar o desempenho de busca/varredura**, **balanceando** uma árvore binária, **evitando** ramos longos e gerando o maior número de ramificações binárias possíveis.

#             7
#           /   \   
#         0       4

Logo, ela é uma BST, com a **diferença** de altura entre a subárvore da direita e da subarvore da esquerda **será sempre 0,1 ou -1**. Fora disto, teremos problemas no balanceamento. Como exemplo, se resultar em 2 ou -2. <br>

- **Primeiro Passo:**
Calcula-se a altura relativa daquele elemento para o lado direito da árvore. Nesse caso, pegamos o nível mais alto do lado direito daquele elemento e subtraimos do nivel do elemnto desejado. <br><br>

- **Segundo Passo:**
Calcula-se a altura relativa daquele elemento para o lado esquerdo da árvore. Nesse caso, pegamos o nível mais alto do lado esquerdo daquele elemento e subtraímos do nível do elemento desejado. <br><br>

- **Terceiro Passo:**
Tendo a altura direita e a esquerda calculada, fazemos a diferença entre elas (direita - esquerda,sempre!). Se o cálculo resultar em 2 ou -2, existe um desbalanceamento e uma rotação será necessára na árvore.

**Exemplo de Desbalanceado:**

#             7
#           /   \
#         4       9
#       /          \
#      0            13
#     / \

**Verificando balanceamente da Raíz:**

- Direita: <br>
Temos: Nivel'raiz'[0] = 7, Nivel[1] = 9. Nivel[2] = 13. Nivel mais longo á direita é o 13, que esta no nível 2, logo, nivel = 2 - raiz = nivel = 0, resulta em 2.<br><br>

- Esquerda: <br>
Temos: Nivel'raiz'[0] = 7, Nivel[1] = 4, Nivel[2] = 0, Nivel[3] = 8 e 5. Nivel mais longo á esquerda é o 8 e 5, que esta no nível 3, logo, nivel = 3 - raiz = nivel = 0, resulta em 3.<br><br>

- Passo 3: direita - esquerda = 2 - 3 = 1.<br>

Logo, com o resultado **1**, esta dentro da faixa aceita que é de 0,1 ou -1, então não há problemas e esta balanceada a **raíz**, do ponto de vista da raiz.<br><br>

**OBS: Precisamso observar se está baçanceada as raízes dos outros nós, dos demais nívels.**<br><br>

```python
# Arvore BInária AVL:
class BST(object):
  def __init__(self, dado): #Método de inicialização e criações de variáveis vazias
    self.dado = dado
    self.esquerda = None
    self.direita = None
    self.altura = 1 # Para calcular a altura (0,1 ou -1) para mensurar o balanceamento.


class AVL(object):
  def inserir(self, root, dado):
    # Passo 1 - BST convencional
    if not root: # Verifica se é a raiz ou não, caso não seja...
      return BST(dado)
    elif dado < root.dado:
      root.esquerda = self.inserir(root.esquerda,dado) # recursividade
    else:
      root.direita = self.inserir(root.direta, dado) # recursividade

    # Passo 2 - Atualiza a Altura do Nó Pai
    # Para verificar se precisa do passo 3 que é balancear
    # calcula apenas a altura da raiz ou nó(raiz que estamos), emr elaçaõ ao ramo  mais longo
    root.altura = 1 + max(self.getAltura(root.esquerda),
                          self.getAltura(root.direita))

    # Passo 3 - Balanceamento
    balanceamento = self.getBalanceamento(root) # Para pegar a variável

    # Passo 4 - Se tiver desbalanceada

    # Direita:
    if balanceamento > 1 and dado < root.direita.dado:
      return self.direitaRotacao(root)

    # Dupla -  Direita e Esquerda:
    if balanceamento < -1 and dado < root.direita.dado:
      root.direita = self.direitaRotacao(root.direita)
      return self.esquerdaRotacao(root) # **



    # Esquerda:
    if balanceamento > -1 and dado < root.esquerda.dado:
      return self.esquerdaRotacao(root)

    # Dupla -  Esquerda e Direita:
    if balanceamento > 1 and dado < root.esquerda.dado:
      root.esquerda = self.esquerdaRotacao(root.esquerda)
      return self.direitaRotacao(root) # **

    return root

  def esquerdaRotacao(self, z):
    y = z.direita
    t2 = y.esquerda

    #rotação:
    y.esquerda = z
    z.direita = t2

    # Atualiza os valores das alturas
    z.altura = 1 + max(self.getAltura(z.esquerda), # tavez tenha que troca direita - esquerda
                       self.getAltura(z.direita))
    y.altura = 1 + max(self.getAltura(y.esquerda),
                       self.getAltura(y.direita))

    return y

  def direitaRotacao(self, z):
    y = z.esquerda
    t2 = y.direita

    #rotação:
    y.direita = z
    z.esquerda = t2

    # Atualiza os valores das alturas
    z.altura = 1 + max(self.getAltura(z.esquerda),
                       self.getAltura(z.direita))
    y.altura = 1 + max(self.getAltura(y.esquerda),
                       self.getAltura(y.direita))

    return y




  def getBalanceamento(self, root):
    if not root:
      return 0
    return self.getAltura(root.direita) - self.getAltura(root.esquerda)  # Subtração das alturas


# Main:

Teste = AVL()
root = None

# Inserindo de forma dirata
root = Teste.inserir(root, 10)
root = Teste.inserir(root, 20)
root = Teste.inserir(root, 30)
root = Teste.inserir(root, 40)
root = Teste.inserir(root, 50)
root = Teste.inserir(root, 25)

Teste.emOrdem(root)


# Modelo:
#
#
#                 30
#                /  \
#              20    40
#             / \    / \
#           10  25      50
```
