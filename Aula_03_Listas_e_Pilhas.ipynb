#Aula 03 - Listas e Pilhas.

Implementações com _arrays dinâmicos_(listas em Python)

-_"lista, pilha, array, etc"**.append(variável)**_: Adiciona um item no final da lista, pilha, array, etc.<br><br>
-_"lista, pilha, array, etc"**.pop()**_:  Remove item que esta no final da lista, pilha, array, etc. Também é posivel indicar um indice para ser excluido da lista, exemplo: **lista.pop(2)**, e removerá o item do indice 2 da lista.<br><br>
-_"lista, pilha, array, etc"**.push(variável)**_: Adiciona um item no local indica pelo indice da lista, pilha, array, etc.<br><br>
-_"lista, pilha, array, etc"**.reverse(variável)**_: Inverte as posições da lista, invertendo do final para o começo.<br><br>
-_"lista, pilha, array, etc"**.insert(indice, item)**_: Adciona um item na posição que desejar em uma lista, pilha, array, etc.

```python
# Código da Pilhas

def push(pilha, top, tamanho, dado): # pilha criada a variável que iria receber o dado,top já foi indicado, tamanho já foi declarado, dado foi incerido
  if len(pilha) == tamanho:
    print("Pilha Cheia! Impossível inserir")
  else:
    pilha.insert(top, dado) #metodo insert, para inserir o dado na pilha no inicio(top, noindice = 0)
    top +=1 # faz com que o valor do dado não seja substitiodo e sim incrementado, adicionado. Quando incrementa ela passa de indice 0 para indice 1,2,3,etc
    return pilha, top

def pop(pilha, top):
  if len(pilha) == 0:
    print("Pilha(lista) vazia! Impossível remover.")
  else:
    del pilha[top] #metodo deletar, na posição indicada.
    top -= 1 # faz com que o valor do dado na pocição top (indice zero, no início) seja removido(deletado).
    return pilha, top




# Main:
top = 0
pilha = []
tamanho = 5

while True:
  print('\n1 - Inserir na Pilha')
  print('2 - Remover da Pilha')
  print('3 - Listar a Pilha')
  print('4 -  Sair')

  op = int(input("Escolha uma opção: "))

  if op == 1:
    dado = int(input('Qual número deseja inserir? '))
    push(pilha, top, tamanho, dado) #função para empilhar

  elif op == 2:
    pop(pilha, top) #função para desempilhar

  elif op == 3:
    for item in pilha: # varrenado os iténs na pilha (lista)
      print(item, end=' - ')

  elif op == 4:
    print('Encerrando...')
    break

  else:
    print('Selecione outra opção...\n')
```

```python
# Código da FILA


def push(fila, fim, tamanho, dado): # pilha criada a variável que iria receber o dado,fim(fila[5]) já foi indicado, tamanho já foi declarado, dado foi incerido
  if len(fila) == tamanho:
    print("Pilha Cheia! Impossível inserir")
  else:
    fila.insert(fim,dado) #metodo insert, para inserir o dado na pilha no final(fim, noindice = 5)
    fim -= 1 # faz com que o valor do dado não seja substitiodo e sim incrementado, adicionado na posição anterior a "5" ultima. 5,4,3 ..
    return fila, fim

def pop(fila, top):
  if len(fila) == 0:
    print("Fila(lista) vazia! Impossível remover.")
  else:
    del fila[top] #metodo deletar, na posição indicada top = indice 0 (top[0]).
    top += 1 # faz com que o valor do dado na pocição top (indice zero, no início) seja removido(deletado).
    return fila, top




# Main:
top = 0
fim = 5 # adicionaod com ultimo itém
fila = []
tamanho = 5

while True:
  print('\n1 - Inserir na Fila')
  print('2 - Remover da Fila')
  print('3 - Listar a Fila')
  print('4 - Sair')

  op = int(input("Escolha uma opção: "))

  if op == 1:
    dado = int(input('Qual número deseja inserir? '))
    push(fila, fim, tamanho, dado) #função para empilhar

  elif op == 2:
    pop(fila, top) #função para desempilhar, pelo top da fila.

  elif op == 3:
    for item in fila: # varrenado os iténs na pilha (lista)
      print(item, end=' - ')

  elif op == 4:
    print('Encerrando...')
    break

  else:
    print('Selecione outra opção...\n')
```
